```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Iris_Data.csv',index_col=0)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df.drop('species',axis=1))
```

```python
scaler_features=scaler.transform(df.drop('species',axis=1))
```

```python
df_new=pd.DataFrame(scaler_features,columns=df.columns[:-1])
```

```python
df_new.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(scaler_features,df['species'],test_size=0.30)
```

```python
from sklearn.naive_bayes import GaussianNB,CategoricalNB,BernoulliNB
```

```python
classifier=GaussianNB()
```

```python
classifier.fit(X_train,y_train)
```

```python
pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
 
```

```python
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
error_list=[]
for i in range(1,40):
  classifier=GaussianNB()
  classifier.fit(X_train,y_train)
  pred_i=classifier.predict(X_test)
  error_list.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_list,color="blue",linestyle="dashed",marker='o',markerfacecolor='red',markersize=11)
plt.title("dtyfugk")
plt.xlabel("hello")
plt.ylabel("hai")
```
