```python
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
webpage = requests.get('https://www.flipkart.com/mens-footwear/sports-shoes/pr?sid=osp,cil,1cu&otracker=nmenu_sub_Men_0_Sports%20Shoes')
```

```python
soup = BeautifulSoup(webpage.content,"html.parser")
```

```python
soup
```

```python
print(soup.prettify())
```

```python
name=soup.find_all('a',class_="IRpwTa")
```

```python
name
```

```python
Product_Name=[]
for i in range(0,len(name)):
  Product_Name.append(name[i].get_text())
```

```python
Product_Name
```

```python
Price=soup.find_all('div',class_="_30jeq3")
```

```python
Price
```

```python
Priceofp=[]
for i in range(0,len(Price)):
  Priceofp.append(Price[i].get_text())
```
